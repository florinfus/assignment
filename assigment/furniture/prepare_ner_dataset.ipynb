```python
def prepare_dataset(products_file, fulltext_file, token_list_len):
    products_df = pd.read_csv(products_file)
    fulltext_df = pd.read_csv(fulltext_file)
    fulltext_df = pd.DataFrame(set(fulltext_df['0']), columns=['name'])
    fulltext_df = fulltext_df[['\\n' not in item for item in fulltext_df['name'].tolist()]]
    fulltext_df.update(fulltext_df.name.apply(lambda x: x[:100]))
    df = pd.DataFrame(
        set(products_df['name']).union(set(fulltext_df['name'])),
        columns=['tokens'])
    df = df.astype(str)
    df['ner_tags'] = df['tokens'].apply(
        lambda x: 'I-LOC' if x in pd.read_csv(products_file)['name'].tolist() else 'O'
    )
    df = df.groupby(df.index//token_list_len)[['tokens', 'ner_tags']].agg(list)
    return df
```

```python
products_file = 'results_products/bydesignmodern.com.csv'
fulltext_file = 'results_fulltext_crawler/bydesignmodern.com.csv'
token_list_len = 16
prepare_dataset(products_file, fulltext_file, token_list_len)
```

```python
df_list = []
for file in products_files[3:]:
    products_file = products_dir + file
    fulltext_file = fulltext_dir + file
    print(products_file, fulltext_file)
    df_list.append(prepare_dataset(products_file, fulltext_file, token_list_len))

train_df = pd.concat(df_list, ignore_index=True)
train_df
```
